```python
#in the above example cust iD is uniques
#but at times cust ID is not unique and repeats itself in the data then we have to use group by function 
#.groupby
#what is the average quantity  sold by region? 
#first we use the .groupby function and then the .agg method
data.groupby('Region',as_index=False).agg({'Quantity':np.mean})


#as_index=false will convert the output into a dataframe.
#.agg accepts aggregation for summarization as its built on numpy
#agg is same as aggregate . Its callable is passed the columns of the data frame one at a time
#agg is used to give the summaries sich as mean ,sum,max,min...
#also we can give it as "mean"...instead of using np.mean..it gives the same output
```

```python
data.groupby('Region').agg({'Quantity':np.mean})

#as you see the column region becomes the row index which is not desirable
#if we dont use as_index is equal to false, so its always best to used as_index=false.
```

```python
#note that we always dont have to pass it as np. but we can also pass as a string
data.groupby('Region',as_index=False).agg({'Quantity':'mean'})

#Note:here we are passing the opertaion in a dictinary where column names becomes the keys and the
#mathematical operations become the values for the dictionary
```

```python
#what is the total sales by category? we can use 'sum' or np.sum

data.groupby('Category',as_index=False).agg({'Sales':'sum'})
```

```python
#multi level coulmn indexing
#we want to sales and profit together by category
#and we want to look at the maximum sales and the total sales by category at the same time in the output
data.groupby('Category')[['Sales','Profit']].agg(['max','sum'])

#note that after grpby category we use two square brackets
#the first bracket is the indexer and second bracket is select coulmns using list wrapper
```

```python
data.groupby('Segment',as_index=False)[['Sales','Profit']].agg({'Sales':np.sum, 'Profit':np.max})  #if we use as_index
```

```python
data.groupby('Segment')[['Sales','Profit']].agg({'Sales':np.sum, 'Profit':np.max}) #without as_index

#coulumn segement becomes the row index as we didnt specify as index
```

```python
#whats is the average discountgiven  by segment
data.groupby('Segment').agg({"Discount":"mean"})
```

```python
#polishing the output

#when this output will be used dynamically#
#like merging this output with the original data to compare the average discount values by segmentwith the overall values
#there will be a problem if the column name is discount and the original data also has the column discount
data.groupby('Segment',as_index=False).agg({"Discount":np.mean}).rename(columns={'Discount':"Average Discount"})
```

```python
#also sort the output #we have lot of values in our output then its better to sort it in descending order
#above there is only three outputs so its okay. but in realtime scenario it will be a bit 
#we use .sort_values
data.groupby('Segment',as_index=False).agg({'Discount':np.mean}).rename(columns={'Discount':'Average Discount'}).sort_values('Average Discount',ascending=False)
```

```python
#which segment of customers are most profitable
data.groupby('Segment',as_index=False).agg({'Profit':np.mean}).sort_values('Profit',ascending=False)
```

```python
#what are the top 5 categorythat give maximum profit?
data.groupby('Category',as_index=False).agg({'Profit':np.sum}).sort_values('Profit',ascending=False).head(5)
```

```python
import pandas as pd
from pandas import DataFrame

#Merging Data Frame'
#Mixed data types -in different data frames there might be different column names
#its a common practice to source the data from different sources so concating is not completly possible
#such cases we merge the two data frames
df1=DataFrame({'CustomerID':[1,2,3,4,5,6],'Product':['Television','Television','Television','Earphone','Earphone','Earphone']})
df2=DataFrame({'CustomerID':[2,4,6],'State':['Texas','Texas','Seattle']})
```

```python
print(df1)
```

```python
print(df2)
```

```python
#A common key is used to merge two dataframe
#Concept is similar to how joins work in SQL
#OUTER MERGE
pd.merge(df1,df2,how='outer',on='CustomerID')
```

```python
#INNER MERGE
pd.merge(df1,df2,how='inner',on='CustomerID')
```

```python
#left join & right join are most frequently used joins in data science
#left join
pd.merge(df1,df2,how='left',on='CustomerID')

```

```python
#right join
pd.merge(df1,df2,how='right',on='CustomerID')
```

```python
df1=DataFrame({'Customerid':[1,2,3,4,5,6],"Product":['Toaster','Toaster','Toaster','Radio','Radio','Radio']})
df2=DataFrame({'CustomerID':[2,4,6],'State':['Alabama','Alabama','Ohio']})
```

```python
df1
```

```python
df2
```

```python
#if you see above we have customer id meantioned with a different way one with caps and another without,
#so we need to merge it using left_on and right on
pd.merge(df1,df2,how='inner',left_on='Customerid',right_on='CustomerID')
```

```python
#we can use the drop method to clean one of the customer id
pd.merge(df1,df2,how='inner',left_on='Customerid',right_on='CustomerID').drop('CustomerID',axis=1)
#remeber axis =1 stands for the entire column
```
